```python
# Hàm tính hiệu của hai số nguyên và in kết quả
def hieu(a, b):
    ket_qua = a - b
    print("Hiệu của hai số là:", ket_qua)

# Hàm tính tích của hai số và trả về kết quả
def tich(a, b):
    return a * b

# Chương trình chính
# Nhập dữ liệu từ bàn phím
so1 = int(input("Nhập số thứ nhất: "))
so2 = int(input("Nhập số thứ hai: "))

# Gọi hàm tính hiệu
hieu(so1, so2)

# Gọi hàm tính tích và nhận kết quả trả về
ket_qua_tich = tich(so1, so2)
print("Tích của hai số là:", ket_qua_tich)
```

```python
# Hàm kiểm tra số nguyên tố
def la_so_nguyen_to(n):
    # Điều kiện loại bỏ các số nhỏ hơn hoặc bằng 1
    if n <= 1:
        return False
    
    # Kiểm tra xem có ước số nào từ 2 đến căn bậc hai của n không
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    
    return True   # Không chia hết cho số nào → là số nguyên tố

# Chương trình chính
# Nhập số từ bàn phím
so = int(input("Nhập một số nguyên: "))

# Gọi hàm và in kết quả
if la_so_nguyen_to(so):
    print(f"{so} là số nguyên tố.")
else:
    print(f"{so} không phải là số nguyên tố.")
```

```python
# Hàm tính giai thừa của một số nguyên dương
def factorial(n):
    giaithua = 1
    for i in range(1, n + 1):
        giaithua *= i
    return giaithua

# Chương trình chính
# Nhập số từ bàn phím
n = int(input("Nhập một số nguyên dương: "))

# Gọi hàm và in kết quả
ket_qua = factorial(n)
print(f"{n}! = {ket_qua}")
```

```python
# Hàm tính diện tích hình tròn
def circle_area(radius):
    pi = 3.14159
    area = pi * radius * radius
    return area

# Chương trình chính
# Nhập bán kính từ bàn phím
r = float(input("Nhập bán kính hình tròn: "))

# Gọi hàm và in kết quả
dien_tich = circle_area(r)
print(f"Diện tích hình tròn là: {dien_tich}")
```

```python
def is_leap_year(year):
    return (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0)

year = int(input("Nhập một năm bất kỳ: "))

print(f"{year} là năm nhuận." if is_leap_year(year) else f"{year} không phải là năm nhuận.")
```

```python
# Hàm tính tổng các chữ số của một số nguyên
def sum_of_digits(n):
    tong = 0
    n = abs(n)  # Đảm bảo với số âm vẫn xử lý đúng

    while n > 0:
        digit = n % 10     # Lấy chữ số cuối
        tong += digit      # Cộng vào tổng
        n //= 10           # Bỏ chữ số cuối

    return tong

# Chương trình chính
number = int(input("Nhập một số nguyên: "))

ket_qua = sum_of_digits(number)
print(f"Tổng các chữ số của {number} là: {ket_qua}")
```

```python
# Hàm đảo ngược số nguyên
def handleReverseTheNumber(n):
    reversed_number = 0
    sign = 1

    # Xử lý trường hợp số âm
    if n < 0:
        sign = -1
        n = abs(n)

    # Đảo ngược số
    while n > 0:
        digit = n % 10               # Lấy chữ số cuối
        reversed_number = reversed_number * 10 + digit  # Thêm vào số đảo
        n //= 10                     # Bỏ chữ số cuối

    return sign * reversed_number

# Chương trình chính
number = int(input("Nhập một số nguyên bất kỳ: "))
result = handleReverseTheNumber(number)

print(f"Số sau khi đảo ngược là: {result}")
```

```python
    # Hàm kiểm tra số hoàn hảo
def is_perfect_number(n):
    if n <= 0:
        return False

    tong = 0

    # Tính tổng các ước số dương nhỏ hơn n
    for i in range(1, n):
        if n % i == 0:
            tong += i

    # Nếu tổng các ước bằng chính nó → số hoàn hảo
    return tong == n


# Chương trình chính
number = int(input("Nhập một số nguyên: "))

if is_perfect_number(number):
    print(f"{number} là số hoàn hảo.")
else:
    print(f"{number} không phải là số hoàn hảo.")
    
```
